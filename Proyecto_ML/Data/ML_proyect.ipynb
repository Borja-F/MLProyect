```python
def label_rank(rank):
    if rank <= 500:
        return 'popular'
    elif rank <= 8000:
        return 'normal'
    else:
        return 'impopular'

df['label'] = df['activity_all_time_rank'].apply(label_rank)
```

```python
df['genres'] = df['genres'].str.replace("' ", "'", regex=False)
df['genres'] = df['genres'].str.replace(" '", "'", regex=False)
df['genres'] = df['genres'].str.replace('[', '', regex=False)
df['genres'] = df['genres'].str.replace(']', '', regex=False)
df['genres'] = df['genres'].str.replace("'", "", regex=False)



split_data_genres = df['genres'].str.split(',', expand=True)


lista_unicos_genres = []
for column in split_data_genres:
    lista_unicos_genres.append(split_data_genres[column].unique())

concat_genres = np.concatenate(lista_unicos_genres)

lista_genres = set(concat_genres)
lista_genres = list(lista_genres)
lista_genres.remove(None)
```

```python
resultados = []
for genre in lista_genres:
    df2 = df[df['genres'].apply(lambda x: genre in x)]
    result = df2['rating'].mean()
    resultados.append(result)
```

```python
my_dict = {key: value for key, value in zip(lista_genres, resultados)}
sorted_dict = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))
```

```python
df['tags'] = df['tags'].str.replace("' ", "'", regex=False)
df['tags'] = df['tags'].str.replace(" '", "'", regex=False)
df['tags'] = df['tags'].str.replace('[', '', regex=False)
df['tags'] = df['tags'].str.replace(']', '', regex=False)
df['tags'] = df['tags'].str.replace("'", "", regex=False)
df['tags'] = df['tags'].str.replace('"', "", regex=False)




split_data_tags = df['tags'].str.split(',', expand=True)


lista_unicos_tags = []
for column in split_data_tags:
    lista_unicos_tags.append(split_data_tags[column].unique())

concat_tags = np.concatenate(lista_unicos_tags)

lista_tags = set(concat_tags)
lista_tags = list(lista_tags)
lista_tags.remove(None)
```

```python
resultados_tags = []
for tags in lista_tags:
    df3 = df[df['tags'].apply(lambda x: tags in x)]
    result = df3['rating'].mean()
    resultados_tags.append(result)
```

```python
my_dict_tags = {key: value for key, value in zip(lista_tags, resultados_tags)}
sorted_dict_tags = dict(sorted(my_dict_tags.items(), key=lambda item: item[1], reverse=True))
```
