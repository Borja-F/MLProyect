```python
def label_rank(rank):
    if rank <= 500:
        return 'popular'
    elif rank <= 8000:
        return 'normal'
    else:
        return 'impopular'
```

```python
df2 = df2.dropna(subset=["original_language"])

df2['label'] = df['activity_all_time_rank'].apply(label_rank)

mask = df2['original_language'].isin(['japanese', 'korean', 'chinese'])
df2 = df2[mask]
```

```python
df['genres'] = df['genres'].apply(lambda x: ast.literal_eval(x))

# Create a list to hold the individual genre names
individual_genres = []

# Extract individual genre names from the lists in the 'genres' column
for genre_list in df['genres']:
    for genre in genre_list:
        individual_genres.extend(genre.strip().split(', '))
```

```python
for column_name in individual_genres:
    df2[column_name] = ""
```

```python
for unique_value in df['genres'].explode().unique():
    df2[unique_value] = df['genres'].apply(lambda x: 1 if unique_value in x else 0)
```

```python
df['tags'] = df['tags'].apply(lambda x: ast.literal_eval(x))

# Create a list to hold the individual genre names
individual_tags = []

# Extract individual genre names from the lists in the 'tags' column
for genre_list in df['tags']:
    for genre in genre_list:
        individual_tags.extend(genre.strip().split(', '))

for column_name in individual_tags:
    df2[column_name] = ""

for unique_value in df['tags'].explode().unique():
    df2[unique_value] = df['tags'].apply(lambda x: 1 if unique_value in x else 0)
```
